```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Training_Data_Set.csv')
df
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.drop(columns='body_type',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Distance '].fillna(df['Distance '].mean(),inplace=True)
df['engine_power'].fillna(df['engine_power'].mean(),inplace=True)
df['door_count'].fillna(df['door_count'].mean(),inplace=True)
df['seat_count'].fillna(df['seat_count'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Maker']=le.fit_transform(df['Maker'])
df['model']=le.fit_transform(df['model'])
df['Location']=le.fit_transform(df['Location'])
df['Owner Type']=le.fit_transform(df['Owner Type'])
df['transmission']=le.fit_transform(df['transmission'])
df['fuel_type']=le.fit_transform(df['fuel_type'])
```

```python
df.dtypes
```

```python
z=(df-df.mean())/df.std()
z
```

```python
df1=df[(z>-2)&(z<2)]
import matplotlib.pyplot as plt
df1.boxplot()
plt.show()
```

```python
z=(df1-df1.mean())/df1.std()
z
```

```python
df2=df1[(z>-2)&(z<2)]
df2.boxplot()
plt.show()
```

```python
z=(df2-df2.mean())/df2.std()
z
```

```python
df3=df2[(z>-2)&(z<2)]
df3.boxplot()
plt.show()
```

```python
df3.isnull().sum()
```

```python
df3.dropna(inplace=True)
```

```python
df3.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
data=pd.DataFrame(ss.fit_transform(df3),columns=df3.columns)
data
```

```python
data.columns
```

```python
x=data[['Id', 'Maker', 'model', 'Location', 'Distance ', 'Owner Type',
       'manufacture_year', 'Age of car', 'engine_displacement', 'engine_power',
       'Vroom Audit Rating', 'transmission', 'fuel_type','seat_count','door_count']]
y=data['Price']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict=pd.DataFrame(model.predict(x))
y_predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y,y_predict)
mse
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae=mean_absolute_error(y,y_predict)
mae
```
